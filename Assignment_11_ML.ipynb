To calculate the PMF of X, we need to determine the probability of getting each score between 10 and 20.

...
P(X = 20) = (1/44)^20 * (43/44)^0 * (10 choose 10)

Note that (n choose k) represents the number of ways to choose k items from a set of n items, which can be calculated using the binomial coefficient formula.

In summary, the PMF of X, the student's score, can be calculated by determining the probability for each possible score between 10 and 20 using the formulas described above.

### 5. Two independent random variables, X and Y,are given such that 

X~Poisson(α) 

and Y~Poisson(β). 

State a new random variable as Z = X + Y. Find out the PMF of Z.

Ans: When two independent random variables X and Y are both Poisson distributed, their sum Z = X + Y follows a new Poisson distribution.

Given:

X ~ Poisson(α)
Y ~ Poisson(β)

The probability mass function (PMF) of a Poisson-distributed random variable X is given by:

P(X = k) = (e^(-λ) * λ^k) / k!

where λ is the parameter of the Poisson distribution, which is equal to the mean and variance of X.

For Z = X + Y, the PMF of Z can be obtained by convolving the PMFs of X and Y. In other words, we sum the probabilities of all possible values of Z.

Let's calculate the PMF of Z:

P(Z = z) = P(X + Y = z)

Since X and Y are independent, we can write:

P(X + Y = z) = Σ[ P(X = x) * P(Y = z - x) ] for all x

where the summation is taken over all possible values of x that yield z when summed with another value from Y.

For a Poisson distribution, the PMF for X is:

P(X = x) = (e^(-α) * α^x) / x!

And for Y:

P(Y = y) = (e^(-β) * β^y) / y!

Now, let's find the PMF for Z:

P(Z = z) = Σ[ P(X = x) * P(Y = z - x) ] for all x

P(Z = z) = Σ[ (e^(-α) * α^x) / x! * (e^(-β) * β^(z - x)) / (z - x)! ] for all x

Since X and Y are independent, we can separate the terms:

P(Z = z) = (e^(-α) * e^(-β)) * Σ[ (α^x * β^(z - x)) / (x! * (z - x)!) ] for all x

Now, the summation part may look familiar as it's the expansion of the binomial coefficient (a + b)^z. Therefore, we can rewrite it as:

P(Z = z) = (e^(-α) * e^(-β)) * [α^z / z! * Σ[ (z choose x) * β^(z - x) ] for all x

where (z choose x) represents the binomial coefficient.

The formula for the binomial coefficient (n choose k) is:

(n choose k) = n! / (k! * (n - k)!)

So, we can now rewrite the PMF of Z:

P(Z = z) = (e^(-α) * e^(-β)) * [α^z / z! * Σ[ (z choose x) * β^(z - x) ] for all x

This formula gives us the PMF of Z in terms of the parameters α and β and the values of z.

### 8. There are 100 men on a ship. If Xi is the ith man&#39;s weight on the ship and Xi&#39;s are independent and identically distributed and EXi = μ = 170 and σXi = σ = 30. Find the probability that the men&#39;s total weight on the ship exceeds 18,000.

There are 100 men on a ship. If Xi is the ith man&#39;s weight on the ship and Xi&#39;s are independent and
identically distributed and EXi = μ = 170 and σXi = σ = 30. Find the probability that the men&#39;s total
weight on the ship exceeds 18,000.

To find the probability that the men's total weight on the ship exceeds 18,000, we can make use of the Central Limit Theorem (CLT) since the Xi's are independent and identically distributed with a known mean (μ) and standard deviation (σ).

According to the CLT, when independent and identically distributed random variables are summed, the resulting distribution approaches a normal distribution as the number of variables increases.

In this case, we have 100 independent and identically distributed random variables (Xi's) representing the weights of the men on the ship. The sum of these random variables, denoted as S, can be considered as the total weight of the men on the ship.

The mean of S is given by E[S] = n * μ = 100 * 170 = 17,000, where n is the number of random variables.

The standard deviation of S is given by σ[S] = √(n) * σ = √(100) * 30 = 300.

We need to calculate the probability that the total weight S exceeds 18,000, which can be written as:

P(S > 18,000) = P((S - E[S]) / σ[S] > (18,000 - E[S]) / σ[S])

Using the properties of the standard normal distribution, we can standardize the above expression:

P(S > 18,000) = P(Z > (18,000 - 17,000) / 300)

P(S > 18,000) = P(Z > 1/3)

To find this probability, we can look up the cumulative distribution function (CDF) of the standard normal distribution or use statistical software. For simplicity, let's use a standard normal distribution table.

Using the standard normal distribution table, we find that P(Z > 1/3) is approximately 0.3707.

Therefore, the probability that the men's total weight on the ship exceeds 18,000 is approximately 0.3707, or 37.07%.
